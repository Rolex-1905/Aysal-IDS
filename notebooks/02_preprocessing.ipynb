```python
import pandas as pd
import numpy as np
import os
import pickle
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

print("Imports done")
```

```python
data_path = r"C:\Users\lenovo\Desktop\Projects\Network-IDS\data"

# Only raw files, skip any processed CSVs
csv_files = [f for f in os.listdir(data_path) 
             if f.endswith('.csv') 
             and not f.startswith('X_') 
             and not f.startswith('y_')]

drop_cols = ['Flow ID', 'Source IP', 'Destination IP', 'Timestamp']

dfs = []
for file in csv_files:
    path = os.path.join(data_path, file)
    print(f"Processing {file}...")
    
    try:
        df_temp = pd.read_csv(path, low_memory=False, encoding='utf-8')
    except UnicodeDecodeError:
        df_temp = pd.read_csv(path, low_memory=False, encoding='latin-1')
        print(f"  (latin-1 encoding used)")
    
    # Strip column name spaces
    df_temp.columns = df_temp.columns.str.strip()
    
    # Drop ID columns
    df_temp.drop(columns=[c for c in drop_cols if c in df_temp.columns], inplace=True)
    
    # Drop non-numeric except Label
    non_numeric = [c for c in df_temp.select_dtypes(exclude=[np.number]).columns 
                   if c != 'Label']
    df_temp.drop(columns=non_numeric, inplace=True)
    
    # Handle inf and NaN column by column (memory efficient)
    for col in df_temp.select_dtypes(include=[np.number]).columns:
        df_temp[col].replace([np.inf, -np.inf], np.nan, inplace=True)
        if df_temp[col].isnull().sum() > 0:
            df_temp[col].fillna(df_temp[col].median(), inplace=True)
    
    # KEY: Sample 30% benign, keep ALL attacks
    benign = df_temp[df_temp['Label'] == 'BENIGN'].sample(frac=0.3, random_state=42)
    attacks = df_temp[df_temp['Label'] != 'BENIGN']
    df_temp = pd.concat([benign, attacks], ignore_index=True)
    
    print(f"  -> {df_temp.shape} | Attacks: {(df_temp['Label'] != 'BENIGN').sum()}")
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)
print(f"\nFinal combined shape: {df.shape}")
print(f"\nLabel distribution:")
print(df['Label'].value_counts())
```

```python
df['Label_Binary'] = df['Label'].apply(lambda x: 0 if x == 'BENIGN' else 1)

le = LabelEncoder()
df['Label_Multi'] = le.fit_transform(df['Label'])

print("Label classes:")
for i, cls in enumerate(le.classes_):
    print(f"  {i} -> {cls}")

print(f"\nBinary distribution:")
print(df['Label_Binary'].value_counts())
```

```python
X = df.drop(columns=['Label', 'Label_Binary', 'Label_Multi'])

# Confirm no non-numeric columns remain
non_numeric = X.select_dtypes(exclude=[np.number]).columns.tolist()
if non_numeric:
    print(f"Dropping remaining non-numeric: {non_numeric}")
    X = X.drop(columns=non_numeric)

y_binary = df['Label_Binary']
y_multi = df['Label_Multi']

print(f"X shape: {X.shape}")
print(f"Any NaN: {X.isnull().sum().sum()}")
print(f"Any inf: {np.isinf(X).sum().sum()}")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print(f"Scaling done")
print(f"Mean (should be ~0): {X_scaled.mean().mean():.4f}")
print(f"Std  (should be ~1): {X_scaled.std().mean():.4f}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_binary, test_size=0.2, random_state=42, stratify=y_binary
)

print(f"Train shape: {X_train.shape}")
print(f"Test shape:  {X_test.shape}")
print(f"\nTrain label distribution:")
print(y_train.value_counts())
print(f"\nTest label distribution:")
print(y_test.value_counts())
```

```python
data_path = r"C:\Users\lenovo\Desktop\Projects\Network-IDS\data"
models_path = r"C:\Users\lenovo\Desktop\Projects\Network-IDS\models"

X_train.to_csv(os.path.join(data_path, "X_train.csv"), index=False)
X_test.to_csv(os.path.join(data_path, "X_test.csv"), index=False)
y_train.to_csv(os.path.join(data_path, "y_train.csv"), index=False)
y_test.to_csv(os.path.join(data_path, "y_test.csv"), index=False)

with open(os.path.join(models_path, "scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)
with open(os.path.join(models_path, "label_encoder.pkl"), "wb") as f:
    pickle.dump(le, f)

print("All saved successfully")
print(f"\ndata/ folder now contains:")
for f in os.listdir(data_path):
    print(f"  {f}")
```
